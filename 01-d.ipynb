# **SIMULACIÓN**

```python
import numpy as np
import simpy 
import random
import matplotlib.pyplot as plt
```

```python
mean_service = data['duracion'].mean()
print("Tiempo de servicio promedio:", mean_service)
```

```python
# Calcular la tasa de llegada (λ)
# Asumiremos que 'hora_llegada' es un valor numérico (por ejemplo, normalizado) y estimaremos el intervalo promedio entre llegadas.
# Para ello, calculamos la diferencia entre clientes consecutivos y luego tomamos el promedio.
data_sorted = data.sort_values(by=['Fecha', 'hora_llegada'])
arrival_intervals = data_sorted['hora_llegada'].diff().dropna()
mean_interval = arrival_intervals.mean()
arrival_rate = 1 / mean_interval  # clientes por unidad de tiempo
print("Intervalo promedio entre llegadas:", mean_interval)
print("Tasa de llegada (λ):", arrival_rate)
```

## **SIMULACIÓN CON SIMPY**

```python
def atender_cliente(env, nombre, operadores, service_time, tiempos_espera):
    """Simula la atención a un cliente."""
    llegada = env.now
    # El cliente solicita atención (acceso a un operador)
    with operadores.request() as req:
        yield req
        # El tiempo de espera es el tiempo transcurrido desde la llegada hasta que comienza la atención
        espera = env.now - llegada
        tiempos_espera.append(espera)
        # Se simula el tiempo de atención, modelado con una distribución exponencial
        yield env.timeout(random.expovariate(1.0/service_time))
        # Fin de la atención (podrías almacenar información adicional si lo requieres)
```

```python
def fuente_clientes(env, arrival_rate, operadores, service_time, tiempos_espera):
    """Genera clientes de acuerdo a una tasa de llegada."""
    i = 0
    while True:
        # Se espera un tiempo (exponencial) entre llegadas
        yield env.timeout(random.expovariate(arrival_rate))
        i += 1
        env.process(atender_cliente(env, f"Cliente {i}", operadores, service_time, tiempos_espera))
```

## **ESCENARIOS DE SUMILACIÓN**

```python
def simular_sistema(num_operadores, sim_time, arrival_rate, service_time):
    env = simpy.Environment()
    operadores = simpy.Resource(env, capacity=num_operadores)
    tiempos_espera = []
    env.process(fuente_clientes(env, arrival_rate, operadores, service_time, tiempos_espera))
    env.run(until=sim_time)
    return tiempos_espera
```

```python
# Definir parámetros de simulación
SIM_TIME = 30  # Tiempo total de simulación un dia
num_operadores_actual = 2  # Escenario actual: número de operadores
num_operadores_alternativo = 3  # Ejemplo: aumentar operadores en horas pico
```

```python
# Simular escenario actual
tiempos_espera_actual = simular_sistema(num_operadores_actual, SIM_TIME, arrival_rate, mean_service)
print("\n--- Escenario Actual ---")
print("Número de clientes atendidos:", len(tiempos_espera_actual))
print("Tiempo promedio de espera:", np.mean(tiempos_espera_actual))
```

```python
# Simular escenario alternativo (más operadores)
tiempos_espera_alternativo = simular_sistema(num_operadores_alternativo, SIM_TIME, arrival_rate, mean_service)
print("\n--- Escenario Alternativo (más operadores) ---")
print("Número de clientes atendidos:", len(tiempos_espera_alternativo))
print("Tiempo promedio de espera:", np.mean(tiempos_espera_alternativo))
```

## **COMPARACIÓN**

```python
plt.figure(figsize=(12,5))
plt.hist(tiempos_espera_actual, bins=20, alpha=0.6, label=f'Operadores = {num_operadores_actual}')
plt.hist(tiempos_espera_alternativo, bins=20, alpha=0.6, label=f'Operadores = {num_operadores_alternativo}')
plt.xlabel("Tiempo de Espera")
plt.ylabel("Frecuencia")
plt.title("Comparación de Tiempos de Espera en Diferentes Escenarios")
plt.legend()
plt.show()
```
